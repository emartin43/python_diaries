```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
name = input("Enter you name: ")

weight = int(input("Enter your weight in pounds: "))

height = int(input("Enter your height in inches: "))

BMI = (weight * 703) / (height * height)

print("Your BMI is " + str(BMI) + ". Disclaimer: BMI is a simple calculation based on weight and height, and does not take into account factors such as muscle mass or body composition.")

if BMI>0:
    if(BMI<18.5):
        print(name +", you are underwight. This means you may be at risk for malnutrition, osteoporosis, and other health issues. You should consider talking to a doctor or a dietitian about how to gain weight in a healthy way.")
    elif (BMI<=24.9):
        print(name +", you are normal weight. This means you have a healthy weight for your height and may be at lower risk for health problems like heart disease, stroke, and diabetes.")
    elif (BMI<29.9):
        print(name +", you are overweight. This means you may be at higher risk for health problems like high blood pressure, high cholesterol, and heart disease. You may want to consider making lifestyle changes such as increasing physical activity and improving your diet.")
        print(name +", you are obese. This means you may be at higher risk for health problems like sleep apnea, joint pain, and diabetes. You should consider talking to a doctor about creating a weight loss plan that works for you.")
    elif (BMI<39.9):
        print(name +", you are severely obese. This means you may be at higher risk for health problems like heart disease, stroke, and high blood pressure. You should consider talking to a doctor about creating a weight loss plan that works for you.")
    else:
        print(name +", you are morbidly obese.  This means you may be at highest risk for health problems like heart disease, stroke, and diabetes. You should consider talking to a doctor about creating a weight loss plan that works for you and may involve medical intervention.")
else:
    print("Enter valid input")
```

```python
#Create custom palette
custom_palette = ListedColormap(['#86765A', '#C5BFB0', '#D2C6B7', '#E8D8C9', '#F0E6D9', '#F8F1EB'])

# BMI scale ranges and corresponding labels
bmi_ranges = [18.5, 25, 30, 35, 40, 50]
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese', 'Severely Obese', 'Morbidly Obese']

# Create a bar chart of the BMI scale
plt.barh(bmi_labels, bmi_ranges, color= custom_palette.colors)

# Add labels and titles to the chart
plt.xlabel('BMI Range')
plt.ylabel('BMI Value')
plt.title('BMI Scale')

# Display the chart
plt.show()
```
